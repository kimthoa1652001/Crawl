```python
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import csv
import pandas as pd
import re
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome(options=options)
url = "https://link.springer.com/search?dc.creator=%22Thanh%20Duc%20Ngo%22"
driver.get(url)
soup = BeautifulSoup(driver.page_source)
```

```python
category = soup.find_all('a', {"class":"title"})
```

```python
u = "https://link.springer.com"
list_url_category = []
title = []
for link in category:
  title.append(link.get_text())
  list_url_category.append(u+link.get("href"))
for i in list_url_category:
  print(i)
```

```python
df = pd.DataFrame()
href = []
time = []
authors = []
abstract = []
for i in list_url_category:
   driver.get(i)
   soup = BeautifulSoup(driver.page_source)
   href.append(i)
  
   x = soup.find('time')
   try:
    time.append(x.getText())
   except:
     time.append("[]")
  
   y = soup.find_all('p',{"class":"c-article-author-affiliation__authors-list"})
   authors.append(y)

   abstract.append(soup.find('div',{"class":"c-article-section__content"},{"id":"Abs1-content"}).get_text()) 
  
```

```python
df['headline'] = title
df['article_link'] = href
df['date'] = time
df['authors'] = authors
df['abstract'] = abstract
df
```
